```python

# Import Gradio and other necessary libraries
import gradio as gr
from PIL import Image
import numpy as np
from ultralytics import YOLO

# Load the YOLO model with pre-trained weights
model_path = '/content/gdrive/MyDrive/MedicalWasteClassification/runs/classify/train4/weights/best.pt'
model = YOLO(model_path)

# Define the function to make predictions
def make_predictions(image):
    results = model(image)
    names_dict = results[0].names
    probs = results[0].probs.data.tolist()
    predicted_item_class = names_dict[np.argmax(probs)].split('_')[0]
    predicted_category = names_dict[np.argmax(probs)].split('_')[1]
    return predicted_item_class, predicted_category, image

# Function to classify waste from an image
def classify_waste(image):
    predicted_item_class, predicted_category, image = make_predictions(image)
    return f"Predicted Waste Item Class: {predicted_item_class}", f"Predicted Waste Category: {predicted_category}", image

# Gradio Interface for uploading image
image_upload_interface = gr.Interface(
    fn=classify_waste,
    inputs="image",
    outputs=["text", "text", "image"],
    title="Medical Waste Detection and Classification - Image Upload",
    description="Upload an image for waste detection and classification."
)

# Gradio Interface for capturing image
image_capture_interface = gr.Interface(
    fn=classify_waste,
    inputs="image",
    outputs=["text", "text", "image"],
    title="Medical Waste Detection and Classification - Image Capture",
    description="Capture an image for waste detection and classification."
)

# Launch the interfaces with debug=True
image_upload_interface.launch(share=True, debug=True)
image_capture_interface.launch(share=True, debug=True)
```
